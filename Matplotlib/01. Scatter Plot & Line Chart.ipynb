```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### ***Scatter Plot***

```python
pizza_size = np.arange(8,13)
pizza_size
```

```python
pizza_price = np.arange(300,550,50)
pizza_price
```

```python
plt.scatter(pizza_size,pizza_price)
```

***To remove the extra statement we use plt.show()***

```python
plt.scatter(pizza_size,pizza_price)
plt.show()
```

```python
plt.scatter(pizza_size,pizza_price)
plt.title("Pizza Price Chart")
plt.xlabel("Pizza Size in inch")
plt.ylabel("Pizza Price in Taka")
plt.xlim(6,13)
plt.show()
```
