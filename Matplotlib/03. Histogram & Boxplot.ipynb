```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### ***Histogram***

```python
x = np.random.randn(10000)
x
```

```python
plt.hist(x,bins=20,color='red',edgecolor='black',alpha=1)
plt.show()
```

#### ***Boxplot***

***Boxplot Shows:***  
**Minimum – the lowest value (excluding outliers)**     
**1st Quartile (Q1) – 25% of the data fall below this value**     
**Median (Q2) – the middle value of the dataset**     
**3rd Quartile (Q3) – 75% of the data fall below this value**     
**Maximum – the highest value (excluding outliers)**    

***It can also show:***     
**Outliers (unusually high or low values)**     
**Interquartile Range (IQR): The range between Q1 and Q3**  

```python
x =  np.random.randint(100,500,50)
x
```

```python
plt.boxplot(x)
plt.show()
```
