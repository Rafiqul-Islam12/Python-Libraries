```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
months = ['January', 'February', 'March']
sale2021 = [8, 10, 17]
sale2022 = [20, 15, 21]
sale2023 = [5, 18, 22]
```

```python
plt.figure()
plt.plot(months, sale2021, label='Sales in 2021')
plt.plot(months, sale2022, label='Sales in 2022')
plt.plot(months, sale2023, label='Sales in 2023')
plt.legend()
```
