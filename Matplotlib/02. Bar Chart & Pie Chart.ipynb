```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### ***Barchart***

***plt.bar() -> vertical bar chart     
plt.barh() -> horizontal bar chart***

```python
blood_group = ['o+', "AB-", "B-"]
blood_group
```

```python
persons = [20,7,30]
persons
```

```python
plt.bar(blood_group, persons,color="green")
plt.title("Blood Group Campaign")
plt.xlabel("Blood Group")
plt.ylabel("Frequency")
```

```python
plt.barh(blood_group, persons,color="green")
plt.title("Blood Group Campaign")
plt.xlabel("Blood Group")
plt.ylabel("Frequency")
```

```python
years = np.arange(2012,2015)
years
```

```python
populations = [200, 250, 300]
populations
```

```python
plt.bar(years, populations)
```

**For customized value in x axis we use xtricks() function**

```python
plt.bar(years, populations)
plt.xticks(years)
plt.show()
```

#### ***Pie Chart***

```python
plt.pie(persons, labels=blood_group, autopct='%1.2f%%', colors= ['red','green','yellow'])
plt.show()
```
