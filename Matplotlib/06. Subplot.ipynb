```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("hw.csv")
df.head()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(df['Heightinch'], df['Weightpound'])
axes[1][0].plot(df['Weightpound'], df['Heightinch'],color='pink')
axes[0][1].scatter(df['Heightinch'], df['Weightpound'],color='red',edgecolor='black')
axes[1][1].scatter(df['Weightpound'], df['Heightinch'],color='green',edgecolor='pink')

fig.suptitle("Main Graph")
axes[0][0].set_title("Graph-01")
axes[0][1].set_title("Graph-02")
axes[1][0].set_title("Graph-03")
axes[1][1].set_title("Graph-04")
fig.tight_layout()
```
