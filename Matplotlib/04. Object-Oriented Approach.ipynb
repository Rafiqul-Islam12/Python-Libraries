```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### ***Object Oriented Method***
**Add axes to the figure**                  
                       ***add_axes([left, bottom, width, height])***
 
**left, bottom, width, height are all in percentage of the canvas**  
**left is the distance from the left edge of the canvas**   
**bottom is the distance from the bottom edge of the canvas**    
**width is the width of the canvas**   
**height is the height of the canvas**   

```python
x = np.arange(0,10)
x
```

```python
y = x**2
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.8,0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.8,0.8])
axes.plot(x,y)

axes1 = fig.add_axes([0.1,0.4,0.4,0.3])
axes1.scatter(x,y)

axes2 = fig.add_axes([0.48, 0.01, 0.3, 0.3])
axes2.plot(y,x)

axes.set_title('Larger Plot')
axes1.set_title('Scatter Plot')
axes2.set_title('y vs x Plot')
```
