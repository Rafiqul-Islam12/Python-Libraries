```python
import pandas as pd
import numpy as np
```

#### ***Generating Random DataFrame***

```python
data = np.random.rand(10, 3)
for col in range(data.shape[1]):
    null_indices = np.random.choice(10, 3, replace=False)
    data[null_indices, col] = np.nan
df = pd.DataFrame(data, columns=['Column_1', 'Column_2', 'Column_3'])
```

```python
df
```

***checking with pd.isna()***

```python
pd.isna(df['Column_1'])
```

```python
pd.isna(df)
```

***Checking with pd.notna()***

```python
pd.notna(df['Column_1'])
```

```python
pd.notna(df)
```

***checking with isnull().sum()***

```python
df.isnull().sum()
```

***checking with notnull().sum()***

```python
df.notnull().sum()
```

```python
df.info()
```

#### ***Handling NaN Values***

```python
df1 = df.copy()
df1.head()
```

```python
df1.fillna(
{
    'Column_1': 0,
    'Column_2' : 1,
    'Column_3': 2
})
```

```python
df.dropna(how='any')
```

```python
df.dropna(how='all')
```
