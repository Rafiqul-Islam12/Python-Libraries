```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/rashakil-ds/Public-Datasets/main/Screen%20Time%20Data.csv'
df = pd.read_csv(url)
df.head()
```

#### ***1.1 Delete Columns Using 'del' Statement***

```python
del df['index']
```

```python
df.head()
```

#### ***1.2 Delete Columns Using 'drop' method***

```python
#df = df.drop('index', axis=1)
df.drop('Date', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=['Week Day', 'Social Networking'], inplace=True)
df.head()
```

#### ***1.3 Drop the Rows in Pandas***

```python
df.drop([0,1]).head()
```

```python
df = df[df['Productivity']>15]
df.head()
```

#### ***2. Rename the columns name***

```python
df = df.rename(columns={'Productivity': 'productivity', 'Yoga': 'yoga'})
df.head()
```
